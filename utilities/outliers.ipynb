```python
import pandas as pd
import numpy as np
import os
import seaborn as sns


path = 'C:\\Users\\davide\\Desktop'
file = 'task_3trialtypes_rt_DIFF_from_Jasp.csv'
df = pd.read_csv(os.path.join(path,file))
```

```python
sns.boxplot(data=df, palette="Set2", width = .18)
sns.swarmplot(data=df, palette="Set2", size=1.5, edgecolor="black", linewidth=.04)
```

```python
df_copy = df.copy()
for col in df.columns:
    # Computing IQR
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_lim = Q1 - 1.5 * IQR
    upper_lim = Q3 + 1.5 * IQR
    filtered = (df[col] > lower_lim) & (df[col] < upper_lim) # Filtering Values between Q1-1.5IQR and Q3+1.5IQR
 
    df[col] = df[col][filtered]

sns.boxplot(data=df, palette="Set2", width = .18)
sns.swarmplot(data=df, palette="Set2", size=1.5, edgecolor="black", linewidth=.04)


```

```python
df.to_csv('df_filtered.csv')
```
